### Choose one variable and plot that variable four different ways.

### Histograms
#### Pro's
-  Viewers can easily compare data.
-  High consistency having equal intervals.
-  Work well with data with a large range.

#### Con's
-  Near impossible to get exact input data points.
-  Not good for multiple categories.
-  Used only for numerical data.

```python
histogram = sns.distplot(dropna_forty, kde=True, bins=20, rug=False);
```

### Box Plots
#### Pro's
-  Good for summarizing large sets of data.
-  Displays the distribution and range on a number line
-  Easy to see outliers and skew

#### Con's
-  Can't determine mean and mode in a boxplot.
-  Original data is not clear.


```python
sns.boxplot(dropna_forty);
```

```python
sns.stripplot(dropna_forty);
```

### Violin Plots
#### Pro's
-  Displays distribution and probability density of a numerical variable.
-  Violin plots are a combination of a Box Plot and a Density Plot that is used to display distribution
-  Displays summary statistics
-  Visualizes the entire data distribution

#### Con's
-  Can be hard to interpret for people unfamiliar.

```python
sns.violinplot(dropna_forty);
```

### Scatter plots
